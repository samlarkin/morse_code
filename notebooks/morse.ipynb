```
import string
import yaml
import json
lettersS = string.ascii_lowercase
digitsS = string.digits
```

```
lettersL = list(lettersS)
digitsL = list(digitsS)
```

```
characters = lettersL + digitsL
```

```
morse_letters = [
    '.-',
    '-...',
    '-.-.',
    '-..',
    '.',
    '..-.',
    '--.',
    '....',
    '..',
    '.---',
    '-.-',
    '.-..',
    '--',
    '-.',
    '---',
    '.--.',
    '--.-',
    '.-.',
    '...',
    '-',
    '..-',
    '...-',
    '.--',
    '-..-',
    '-.--',
    '--..'
]
```

```
letters_dictionary = {}
for index in range(len(lettersL)):
    letters_dictionary.update({lettersL[index] : morse_letters[index]})
```

```
#letters_dictionary
```

```
morse_digits = [
    '-----',
    '.----',
    '..---',
    '...--',
    '....-',
    '.....',
    '-....',
    '--...',
    '---..',
    '----.'
]
```

```
digits_dictionary = {}
for index in range(len(digitsL)):
    digits_dictionary.update({digitsL[index] : morse_digits[index]})
```

```
#digits_dictionary
```

```
punctuation_dictionary = {
    '.' : '.-.-.-',
    ',' : '--..--',
    '?' : '..--..',
    "'" : '.----.',
    '!' : '-.-.--',
    '/' : '-..-.',
    '(' : '-.--.',
    ')' : '-.--.-',
    '&' : '.-...',
    ':' : '---...',
    ';' : '-.-.-.',
    '=' : '-...-',
    '+' : '.-.-.',
    '-' : '-....-',
    '_' : '..--.-',
    '"' : '.-..-.',
    '$' : '..-..-',
    '@' : '.--.-.'
}
```

```
morse_prosigns_dictionary = {
    'END_OF_WORK' : '...-.-',
    'ERROR' : '........',
    # k == 'INVITATION_TO_TRANSMIT' : '-.-',
    'STARTING_SIGNAL' : '-.-.-',
    'NEW_PAGE_SIGNAL' : '.-.-.',
    'UNDERSTOOD' : '...-.',
    # & == 'WAIT' : '.-...'
}
```

```
dictionary = {}
dictionary.update(letters_dictionary)
dictionary.update(digits_dictionary)
dictionary.update({' ' : ' '})
dictionary.update(punctuation_dictionary)
dictionary.update(morse_prosigns_dictionary)
```

```
#dictionary
with open('/content/drive/My Drive/Colab Notebooks/morse-code/morse_dictionary.yaml', 'w+') as f:
    f.write(yaml.dump(dictionary))

with open('/content/drive/My Drive/Colab Notebooks/morse-code/morse_dictionary.json', 'w+') as f:
    f.write(json.dumps(dictionary))
```

```
def encode(message):
    #if not bool((type(message) is type(0)) or (type(message) is type(''))):
    if not (isinstance(message, int) or isinstance(message, str)):
        raise TypeError('encode(message) handles only a message of type: str or int')
    elif isinstance(message, int):
        input_string = str(message)
    else:
        input_string = message
    input_list = list(input_string)
    working_list = []
    count = 0
    for char in input_list:
        count += 1
        if char in dictionary.keys():
            working_list.append(dictionary[char])
            if count < len(input_list):
                working_list.append(' ')
        elif char in list(string.ascii_uppercase):
            char = char.lower()
            working_list.append(dictionary[char])
            if count < len(input_list):
                working_list.append(' ')
        else:
            raise ValueError
    output = ''.join(working_list)
    return output
```

```
message = 'The quick brown fox jumps over the lazy dog'
```

```
encode(message)
```

```
def decode(morse_message):
    input_words = morse_message.split('  ')
    working_list = []
    count = 0
    for word in input_words:
        count += 1
        chars_in_word = word.split(' ')
        for char in chars_in_word:
            if char in dictionary.values():
                for key,value in dictionary.items():
                    if value == char:
                        working_list.append(key)
            elif char == '':
                pass
            else:
                raise ValueError
        if  count < len(input_words):  
            working_list.append(' ')
    output = ''.join(working_list)         
    return output
```

```
morse_message = encode(message)
decode(morse_message)
```
